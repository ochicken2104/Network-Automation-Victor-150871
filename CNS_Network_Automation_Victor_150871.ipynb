### Define the device credentials

### Roles of connect handler
1. Establish SSH connection to network device
2. Automatically handles handshake

**NB** The moment you specify device type connectHandler knows how to
1. Specifically navigate that specific OS
2. Wait for device to finish processing a command

##Gatepass The security gatekeeper
'getpass' is a standard python module used to keep credentials safe.

Problem is when you hardcode pass directly to script e.g "Victor @123" anyone who sees your code has access to your github respiratory

**solution**
'getpass' prompts user for passsword in terminal/console but hides characters as typed

```python
from netmiko import ConnectHandler
#Netmiko, Different vendors
#command-line syntaxes, ConnectHandler acts as universal adapter
import getpass

device = {
    'device_type':'cisco_xe',
    'host': 'sandbox-iosxe-latest-1.cisco.com',
    'username': 'developer',
    'password': 'Ry@n123',
    'port': 22,
}

print(f"Targeting: {device['host']}")
```

Retrieve Information(Show the commands)

Goal
-Establish a connection ans
-Pull the interface status
This is the "Read" part of the automation.


```python
try:
  #Establish SSH connection
  print("Connecting to the device...")
  net_connect = ConnectHandler(**device)

  #Run the show command
  output = net_command = net_connect.send_command('show ip int brief')

  print("\n--- Interface Status ---")
  print(output)

  #Disconnect
  net_connect.disconnect()

  except Exception as e:
    print(f"Error connecting to device: {e}")
```

Pushing Configurations

Automation is particularly spectacular when you want to change settings.

Next let us create a new virtual interface called Loopback 99.

```python
#Re-establish connections for the config change

net_connect = ConnectHandler(**device)

config_commands = [
    'Interface Loopback99',
    'description Configured via Python Netmiko',
    'ip address 192.168.99.1 255.255.255.255',
]

print("Sending configuration commands...")
output = net_connect.send_config_set(config_commands)
print(output)

#Verify the change
verification= net_connect.send_command('show ip int brief | include Loopback99')
print("\nVerification")
print(verification)

net_connect.disconnect
```
